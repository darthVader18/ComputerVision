# CallBack & Trackbars

```python
black_img = np.zeros((512, 512, 3), np.uint8)
cv2.namedWindow("Black Image")

def draw_circle(event, x, y, flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(black_img, (x, y), 50,(255, 50, 70),5)
        

cv2.setMouseCallback('Black Image',draw_circle)

while(1):
    cv2.imshow('Black Image', black_img)
    k = cv2.waitKey(5)
    if k == 27:
        break

cv2.destroyAllWindows()
```

```python
cv2.namedWindow("Image")

cv2.createTrackbar('Channel','Image',0,2,lambda x:print(x))

while(1):
    cv2.imshow('Image',img[:,:,v])
    v = cv2.getTrackbarPos('Channel','Image')
    
    k = cv2.waitKey(5)
    if k == 27:
        break

cv2.destroyAllWindows()    
```

```python
b, g, r = cv2.split(img)
cv2.imshow('b', b)
cv2.imshow('g', g)
cv2.imshow('r', r)
cv2.waitKey( 0 )
cv2.destroyAllWindows()
```
